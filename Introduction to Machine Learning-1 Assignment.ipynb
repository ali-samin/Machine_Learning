#### Q5- What are the main difference between supervised ,unsupervised and  semi-supervised learning.

The main differences between supervised, unsupervised, and semi-supervised learning:

Supervised Learning:
- the algorithm is trained on a dataset that is labeled, meaning each input data point is associated with a corresponding target label or output.
- The goal of supervised learning is to learn a mapping from input variables to output variables based on the given labeled training data.
- During training, the algorithm adjusts its parameters to minimize the difference between its predictions and the actual target values in the training data.
- Supervised learning is commonly used for tasks such as classification and regression.


Unsupervised Learning:
- the algorithm is trained on a dataset that is unlabeled, meaning the input data does not have corresponding target labels.
- The goal of unsupervised learning is to find hidden patterns, structures, or relationships in the data without explicit guidance.
- Unsupervised learning algorithms explore the data and identify inherent structures or clusters based on the underlying patterns present in the data.
- Unsupervised learning is commonly used for tasks such as clustering, dimensionality reduction, and density estimation.


Semi-Supervised Learning:
- Semi-supervised learning lies between supervised and unsupervised learning, where the algorithm is trained on a dataset that contains a small amount of labeled data along with a larger amount of unlabeled data.
- The goal of semi-supervised learning is to leverage the available labeled data to improve the learning process and make better predictions or decisions on the unlabeled data.
- Semi-supervised learning algorithms combine the benefits of supervised and unsupervised learning, using labeled data for guidance while also exploring the structure of the unlabeled data.
- Semi-supervised learning is particularly useful when obtaining labeled data is expensive or time-consuming, as it allows leveraging the abundance of unlabeled data to improve performance.

#### Q6- What is train, test and validation split? Explain the importance of each them.

Training Set:

- The training set is a subset of the dataset used to train the machine learning model. It consists of input data and corresponding target labels (in supervised learning) or just input data (in unsupervised learning).
- The model learns from the training set by adjusting its parameters to minimize the error between its predictions and the actual target values.
- Importance: The training set is crucial for building and fine-tuning the model. It provides the data necessary for the model to learn the underlying patterns and relationships in the data.


Validation Set:
- The validation set is a separate subset of the dataset used to tune the hyperparameters of the model and evaluate its performance during training.
- It helps assess how well the model generalizes to new, unseen data and prevents overfitting by providing an unbiased estimate of the model's performance.
- The validation set is used iteratively during the training process to adjust hyperparameters, such as learning rate, regularization strength, or network architecture, based on the model's performance on the validation set.
- Importance: The validation set helps ensure that the model's performance is optimized and generalizes well to unseen data by providing an independent evaluation during training.


Test Set:
- The test set is another separate subset of the dataset that is used to evaluate the final performance of the trained model after the training and validation stages are completed.
- The test set contains data that the model has not seen during training or validation, allowing for an unbiased assessment of its performance on unseen data.
- Importance: The test set provides an objective measure of the model's generalization performance and helps assess how well it performs in real-world scenarios. It serves as a final check before deploying the model in production.

unsupervised learning techniques are used in anomaly detection by identifying unusual patterns, clusters, or deviations in data without labeled examples of anomalies. Methods include clustering, density estimation, dimensionality reduction, association rule mining, and one-class classification. Anomalies are detected based on their deviation from normal data patterns or distributions.
